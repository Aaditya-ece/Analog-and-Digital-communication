```matlab
%  Create message and carrier

fs = 100000;
t = 0.0: 1/fs: .02;

fc = 5000;
fm = 500;

c= cos(2*pi*fc*t);
m = cos(2*pi*fm*t);

dsb = m.*c;

m_h = imag(hilbert(m));

S_usb = m.*cos(2*pi*fc*t) - m_h.*sin(2*pi*fc*t);
S_lsb = m.*cos(2*pi*fc*t) + m_h.*sin(2*pi*fc*t);

subplot(3, 1, 1)
plot(t, m)
title("message signal")
subplot(3, 1, 2)
plot(t, S_usb)
title("usb signal")
subplot(3, 1, 3)
plot(t, S_lsb)
title("lsb signal")
```

```matlab
N = length(t);
f=  (-N/2: N/2-1)*(fs/N);

usb_fft = fftshift(abs(fft(S_usb)));
lsb_fft = fftshift(abs(fft(S_lsb)));
dsb_fft = fftshift(abs(fft(dsb)));

subplot(3, 1, 1)
plot(f, usb_fft)
xlim([0, 10000])
title("usb_sc spectrum")

subplot(3, 1, 2)
plot(f, lsb_fft)
xlim([0, 10000])
title("lsb_sc spectrum")

subplot(3, 1, 3)
plot(f, dsb_fft)
xlim([0, 10000])
title("dsb_sc spectrum")
```
