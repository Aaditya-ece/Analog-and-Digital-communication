
# <span style="color:rgb(213,80,0)">Generate and Modulate signal</span>

```matlab

clc;
clear;

fs = 10000;
t= 0.0: 1/fs: .05;

Am = 1;
Ac = 1;

fm = 100;
fc = 2000;

B = 5;

sfm= Ac*cos(2*pi*fc*t+ B*sin(2*pi*fm*t));

m = Am*cos(2*pi*fm*t);
```

## FM demod

```matlab

demod = fmdemod(sfm, fc, fs, 500);
figure;
subplot(3, 1, 1);
plot(t, m)
title("message signal");

subplot(3, 1, 2)
plot(t, sfm)
title("Modulated signal")

subplot(3, 1, 3)
plot(t, demod)
title("Demodulated output");
```


## fft verification

```matlab
N = length(t);
f = (-N/2:N/2-1)*(fs/N);

y = fftshift(abs(fft(demod)));

figure;
plot(f, y)
title("FFt of demodulated output");
xlabel("frequency");
ylabel("magnitude");
xlim([-400, 400])
```
